# 2 Plotting a Function
Goal: Write a **function** that will plot $f(x) = x^2$ and $g(x) = \sqrt{x}$ on the same plot and return the $f(x)$ and $g(x)$ values as separate lists. Your plot needs to have the fo/llowing:
- An appropriate figure size (`figsize=(10,8)` is recommended)
- Labeled x and y axis
- A title
- Different colors and linestyles for each function
- A legend
- Your domain should be $x \in [0, 10]$. *Hint: Use `np.linspace` to set a range of values for your functions.*

Your function should return:
- A list of $f(x)$ values
- A list of $g(x)$ values.

```python
def plot_functions():
    # Define the range of x values
    x = np.linspace(0, 10, 100)  # Generates 100 points from 0 to 10

    # Define the functions you want to plot
    y1 = x**2                # Example function 1: sine
    y2 = np.sqrt(x)               # Example function 2: cosine

    # Create the plot
    plt.figure(figsize=(10, 8))
    plt.plot(x, y1, color='blue', linestyle='-', label='x^2')
    plt.plot(x, y2, color='red', linestyle='--', label='sqrt(x)')

    # Customize the plot
    plt.xlabel('x values')
    plt.ylabel('y values')
    plt.title('Plot of x^2 and sqrt(x)')
    plt.legend()  # Add legend to distinguish functions
    plt.grid(True)  # Optional: Add grid for easier visualization

    # Show plot
    plt.show()

    # Return lists of y values for both functions
    return y1.tolist(), y2.tolist()

# Call the function and capture the output lists
y1_values, y2_values = plot_functions()
```

```python
# Import your module and/or your code here
# https://matplotlib.org/stable/gallery/mplot3d/surface3d.html
# This link helped me understand how to plot 3D surfaces and use np.meshgrid to define grid points for plotting.
def plot_wavy_surface():
    # Create a grid of x and y values
    x = np.linspace(-5, 5, 100)
    y = np.linspace(-5, 5, 100)
    X, Y = np.meshgrid(x, y)
    
    # Define a 3D surface as a function of X and Y using sine and cosine
    Z = np.sin(np.sqrt(X**2 + Y**2)) * np.cos(np.sqrt(X**2 + Y**2))
    
    # Create a 3D plot
    fig = plt.figure(figsize=(8, 6))
    ax = fig.add_subplot(111, projection='3d')
    surf = ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')
    
    # Add a color bar, title, axis labels, and legend
    fig.colorbar(surf, ax=ax, shrink=0.5, aspect=5)
    ax.set_title("3D Wavy Surface Plot of Sine and Cosine Functions")
    ax.set_xlabel("X-axis")
    ax.set_ylabel("Y-axis")
    ax.set_zlabel("Z-axis")

    plt.show()

# Uncomment the line below to call the function and generate the plot
plot_wavy_surface()
```
